```python
print('BusinessTravel :',employee['BusinessTravel'].unique())
print('Department :',employee['Department'].unique())
print('EducationField :',employee['EducationField'].unique())
print('Gender :',employee['Gender'].unique())
print('JobRole :',employee['JobRole'].unique())
print('MaritalStatus :',employee['MaritalStatus'].unique())
print('Over18 :',employee['Over18'].unique())
print('OverTime :',employee['OverTime'].unique())
```

```python
cat_cols = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in cat_cols:
    employee[col] = le.fit_transform(employee[col])
    print(f'{col} :', dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
# now that all columns have been converted, we would check for the columns that would not be useful and remove them

print(employee['EmployeeCount'].unique())
print(employee['EmployeeNumber'].unique())
print(employee['EmployeeNumber'].nunique())
print(employee['StandardHours'].nunique())
```

```python
# since EmployeeCount and EmployeeNumber are not useful for our model, we would drop them and even StandardHours has the same value throughout
# also from the EmployeeNumber column we can observe that there are no repeating or duplicate values

df = employee.drop(columns=['EmployeeCount','EmployeeNumber','Over18','StandardHours'])
df
```

```python
x = df.drop(columns='Attrition')
y = df[['Attrition']]

# now we can split our data into train and test and build different models for evaluation
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1234)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model_lr = lr.fit(x_train, y_train)
model_lr
```

```python
y_test['lr_pred'] = lr.predict(x_test)
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, roc_auc_score
print(confusion_matrix(y_test['Attrition'], y_test['lr_pred']))
```

```python
print(classification_report(y_test['Attrition'], y_test['lr_pred']))
```

```python
fpr, tpr, thresholds = roc_curve(y_test['Attrition'], y_test['lr_pred'])
plt.plot(fpr, tpr); plt.xlabel("False Positive Rate"), plt.ylabel('True Positive Rate')
```

#### In the Logistic Regression ROC Curve plot, we are getting a straight line output. This indicates that the model is not good. Also from the confusion matrix it can be observed that one value is predicted correctly but the other is not; even the classification report shows that the model is biased towards predicting one type of value.

### Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model_dt = dt.fit(x_train, y_train)
model_dt
```

```python
y_test['dt_pred'] = model_dt.predict(x_test)
y_test
```

```python
print(confusion_matrix(y_test['Attrition'], y_test['dt_pred']))
```

```python
print(classification_report(y_test['Attrition'], y_test['dt_pred']))
```

```python
fpr, tpr, thresholds = roc_curve(y_test['Attrition'], y_test['dt_pred'])
plt.plot(fpr, tpr); plt.xlabel("False Positive Rate"), plt.ylabel('True Positive Rate')
```

#### In the Decision Tree ROC Curve plot, we are getting a slight deviation around 0.4, so it is a better model than Logistic regression. Also the metrics of classification report and confusion matrix show better results than logistic regression.

### Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500)
model_rf = rf.fit(x_train, y_train)
model_rf
```

```python
y_test['rf_pred'] = model_dt.predict(x_test)
y_test
```

```python
print(confusion_matrix(y_test['Attrition'], y_test['rf_pred']))
```

```python
print(classification_report(y_test['Attrition'], y_test['rf_pred']))
```

```python
fpr, tpr, thresholds = roc_curve(y_test['Attrition'], y_test['rf_pred'])
plt.plot(fpr, tpr); plt.xlabel("False Positive Rate"), plt.ylabel('True Positive Rate')
```

#### In the Random Forest ROC Curve plot, we are getting a the same result as in the Decision Tree above, so even this can be considered as a good output result, if not better.

```python
# auc score
print("Logistic Regression :", roc_auc_score(y_test['Attrition'], y_test['lr_pred']))
print("Decision Tree :", roc_auc_score(y_test['Attrition'], y_test['dt_pred']))
print("Random Forest :", roc_auc_score(y_test['Attrition'], y_test['rf_pred']))
```
